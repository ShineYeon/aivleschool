## 1.데이터 살펴보기

* 주어진 데이터로부터 다양한 조회 방법을 통해 비즈니스 사항을 파악해 봅시다.
* 데이터와 비즈니스를 파악하기 위한 질문을 10개 이상 도출하고, 조회해 봅시다.

```python
#지역 판매량이 높은순으로 출력하기
data.sort_values(by='Sales', ascending=False)
```

```python
#지역별 판매가격의 평균가격
data['Price'].mean()
```

```python
#미국매장의 수 출력
data['US'].value_counts()
```

```python
#도심에 있는 매장 수 출력
data['Urban'].value_counts()
```

```python
#좀 설명 좀 해봐
data.describe()
```

```python
#US/not US별로 Education 집계해줌
data[['US', 'Education']].value_counts()
```

```python
#판매량 높은 순으로, 판매량 같다면 경쟁사 가격 높은 순으로
data.sort_values(by=['Sales', 'CompPrice'], ascending=[False, False])
```

```python
#진열 상태 카운트
data['ShelveLoc'].value_counts()
```

```python
#지역 인구수가 광고 ㅇㅖ산에 영향을 미치는지?
#지역 인구수 출력 광고 예산 출력 (평균값)
data[['Population', 'Advertising']].mean()
```
