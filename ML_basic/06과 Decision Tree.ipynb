* max_depth : 트리의 깊이(크기를 결정)
    - depth : 루트~리프 가장 긴 경로
* min_samples_leaf : leaf 노드의 최소 데이터 건수

### (2) 변수 중요도

```python
# 변수 중요도
print(list(x_train))
print(model.feature_importances_) #변수 중요도를 나타냄.
```

* 변수중요도 그래프 그리기 함수 만들기

```python
def plot_feature_importance(importance, names):#중요도와 이름을 받아서
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    #numpy로 바꾸고

    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)

    # 정렬
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    fi_df.reset_index(drop=True, inplace = True)

    #그림그리기
    plt.figure(figsize=(10,8))
    sns.barplot(x='feature_importance', y='feature_names', data = fi_df)

    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
    plt.grid()

    return fi_df
```

```python
# 첫번째 변수: feature 값들, 두번째변수: feature 이름들
result = plot_feature_importance(model.feature_importances_, list(x_train))

#가장 중요한 변수가 monthlyincome, 두번째가 age, 세번째가 totalWorkingYears이구나.

# monthlyincome이 가장 중요한데 왜 decisiontree에서는 overtime으로 가장 먼저 나누나요?
# overtime이 6위인데도 불구하고??
# 맨처음에 사용한 변수가 overtime
# 왜 overtime? 부모의 불순도를 가장 떨어뜨려주는 기준이 되는 변수가 overtime 이기 때문
# 근데, 모델 전체로 확장하는 순간, 위에 뿐만아니라 밑의 노드들도 봄
# => 이 불순도를 평균으로 계산해서 이 불순도에 가장 영향을 미치는 변수

# 요약하자면, monthlyincome은 모델 전반에서 중요한 변수
# overtime은 가장 처음에 split 할 때 중요한 변수

# 모델 전체로 확장해서 판단한 것이다.~
```

* 그래서 여러분에게, 어떤 고객이 번호이동(이탈)해 가는지 예측 모델링을 의뢰하였습니다.

* 변수설명
    * COLLEGE : 대학 졸업여부
    * INCOME : 연수입
    * OVERAGE : 월평균 초과사용 시간(분)
    * LEFTOVER : 월평균 잔여시간비율(%)
    * HOUSE : 집값
    * HANDSET_PRICE : 스마트폰 가격
    * OVER_15MINS_CALLS_PER_MONTH : 월평균 장기통화(15분이상) 횟수

    * REPORTED_SATISFACTION : 만족도 설문조사 결과
    * REPORTED_USAGE_LEVEL : 사용도 자가진단 결과
    * CONSIDERING_CHANGE_OF_PLAN : 향후 변경계획 설문조사 결과
    * CHURN : 이탈(번호이동) 여부 (Target 변수)


### (1) 데이터 전처리

```python
# 데이터를 불러옵시다.
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/mobile_cust_churn.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

* 변수정리 : 불필요한 변수를 정리합시다.
    * 식별자 : 일련번호, 주민번호, 전화번호, 고객ID, 사번 
    * 시계열 데이터 중 : 어떤 기간동안 거의 변화가 없는 값.(변동이 거의 없는) 값.


```python
drop_cols = ['id']
data.drop(drop_cols, axis = 1, inplace = True )
```

* x, y 분할

```python
target = 'CHURN'
x = data.drop(target, axis = 1)
y = data.loc[:, target]
```

* 가변수화 

```python
dumm_cols = ['REPORTED_SATISFACTION','REPORTED_USAGE_LEVEL','CONSIDERING_CHANGE_OF_PLAN']
x = pd.get_dummies(x, columns = dumm_cols, drop_first = True)
```

* train, val 분할

```python
x_train, x_val, y_train, y_val = train_test_split(x,y,test_size = 0.3)
```

### (2) 모델링
* max_depth를 1~10 사이에서 조정하며 모델을 생성하고 성능을 비교해 봅시다.
* min_samples_leaf 를 20 ~ 300 사이에서 조정하며 모델을 생성하고 성능을 비교해 봅시다.

```python
depth = range(1, 11)
act1 = []
for d in depth:
    m = DecisionTreeClassifier(max_depth=d)
    m.fit(x_train, y_train)
    p = m.predict(x_val)
    act1.append(accuracy_score(y_val, p))
```

```python
leaf_samples = range(20, 301, 10)
act2 = []
for s in leaf_samples:
    m = DecisionTreeClassifier(min_samples_leaf=s)
    m.fit(x_train, y_train)
    p = m.predict(x_val)
    act2.append(accuracy_score(y_val, p))
```

```python
model1 = DecisionTreeClassifier(max_depth=3, min_samples_leaf=100)
model1.fit(x_train, y_train)
pred1 = model1.predict(x_val)
print(classification_report(y_val, pred1))
```

```python
model2 = DecisionTreeClassifier(max_depth=10, min_samples_leaf=100)
model2.fit(x_train, y_train)
pred2 = model2.predict(x_val)
print(classification_report(y_val, pred2))
```

```python
model3 = DecisionTreeClassifier(max_depth=10, min_samples_leaf=50)
model3.fit(x_train, y_train)
pred3 = model3.predict(x_val)
print(classification_report(y_val, pred3))
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.plot(depth , act1, marker='o')
plt.xlabel('depth')
plt.ylabel('accuracy')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(leaf_samples , act2, marker='o')
plt.xlabel('leaf_samples')
plt.ylabel('accuracy')
plt.grid()

# depth의 경우
# 최적이 되는 구간이 있다.
# depth에서 accurary가 증가했다가 감소했다가 다시 증가하지는 않는다.

# leaf_samples는 저 마지막 구간에서 성능이 다시 증가하는 일이 잘 벌어지지 않는다.

# 적절한 지점이 있음. => 찾아봐야 함. 튜닝은 적절한 지점을 찾는 것!!
```
