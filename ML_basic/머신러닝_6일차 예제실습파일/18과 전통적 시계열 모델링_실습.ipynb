# 1.환경준비

## (1) 라이브러리 로딩

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as spst
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from sklearn.metrics import *
from sklearn.model_selection import train_test_split

import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.filterwarnings(action='ignore')
warnings.simplefilter('ignore', ConvergenceWarning)
```

## (2) 함수 생성 

* 결과 시각화

```python
def plot_model_result(y_train, y_val, pred) :
    pred = pd.Series(pred, index = y_val.index)

    # 전체 시각화
    plt.figure(figsize = (20,12))
    plt.subplot(2,1,1)
    plt.plot(y_train, label = 'train')
    plt.plot(y_val, label = 'val')
    plt.plot(pred, label = 'pred')
    plt.legend()
    plt.grid()

    plt.subplot(2,1,2)
    plt.plot(y_val, label = 'val')
    plt.plot(pred, label = 'pred')
    plt.legend()
    plt.grid()

    plt.show()
```

* 잔차분석

```python
def residual_diag(residuals, lags = 30) :
    print('* 정규성 검정(> 0.05) : ', round(spst.shapiro(residuals)[1],5))
    print('* 정상성 검정(< 0.05) : ', round(sm.tsa.stattools.adfuller(residuals)[1],5))
    print('* 자기상관성 확인(ACF, PACF)')
    fig,ax = plt.subplots(1,2, figsize = (15,5))
    plot_acf(residuals, lags = lags, ax = ax[0])
    plot_pacf(residuals, lags = lags, ax = ax[1])
    plt.show()
```

    * item : 상품코드 (여기서는 한가지 상품만 있음)
    * sales : A유통회사 a 매장 판매량 ==> target
    * tot_sales : A유통회사 전체 판매량
    * comp_sales : 인근에 위치한 B유통회사 b 매장 판매량

```python
plt.figure(figsize = (20,8))
plt.plot(data['sales'])
plt.grid()
plt.show()

# 흐름이 물결 흐름
```

# 2.기본 전처리

## (1) y 만들기

* 1일 후 수요량을 예측하려면, y를 어떻게 만들어야 할까요?

```python
data['y'] = data['sales'].shift(-1) #다음날예측해야하니까 하나를 땡겨올림
display(data.head())
display(data.tail())
```

```python
# 제일 마지막 행은 삭제
data.dropna(axis = 0, inplace = True)
data.tail()

# 2/27까지 데이터가 있네요~
```

## (2) Feature Engineering
* 필요한 변수들을 도출해 봅시다.

```python
# SARIMAX 할 때 필요 (실전에선 엄청 중요한데 지금은 안할게요~)
```

## (3) 데이터 분할

### 1) x, y 나누기

```python
target = 'y'

x = data.drop([target, 'date'], axis = 1) #제거할 때, date도 제거
y = data.loc[:, target]
```

### 2) train, val 분할
* 1회 분할 : train_test_split( x, y, test_size= , shuffle = False) 
    * test_size : 소수 - 비율, 자연수 - 갯수
    * shuffle = False : 섞지 말고 데이터 끝에서 test_size 만큼 자르기
* k-fold cross validation for Time Seires : TimeSeriesSplit

```python
# 여기서는 1회 분할로 진행합니다.
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 30, shuffle = False)
```

# 3.모델링1 : ARIMA

## (1) y 값 살펴보기

```python
residual_diag(y_train)

# 정규성 : x
# 정상성 : x
# 자기상관성도 x
# 왜 이렇게 나오냐면 오차값이 아니니까!!!!!!!!
# 정규성>0.05, 정상성<0.05, 자기상관성 파란색범위 안은 오차를 위한(화이트노이즈) 기준임!!!!!
# 저 세개를 다 만족해야 화이트노이즈 => 모델링이 완성된 것
```

## (2) 모델링 : 초기모델

* p, d, q 값을 어떻게 정해야 할까요?
* AR의 p 차수와 MA q 차수 모두 값이 필요해 보입니다. 일단 1, 1을 지정합시다.

### 1) 학습

* sm.tsa.SARIMAX(train, order=(p,d,q)).fit()
    * 모델 선언시 train이 포함
    * .fit()으로 학습.

```python
m1 = sm.tsa.SARIMAX(y_train, order=(1, 0, 1)).fit()
```

```python
m2 = sm.tsa.SARIMAX(y_train, order=(1, 1, 1)).fit()
```

### 2) 평가

#### ① 잔차진단

* 모델.resid : 잔차를 뽑을 수 있습니다.
* 위에서 만든 함수 residual_diag 를 사용하여 잔차진단을 해 봅시다.

```python
m1.predict()
# y_train 갖고 예측한 값
# 이 값에서 실제값을 뺀 게 잔차(.resid)
```

```python
residual1 = m1.resid
residual_diag(residual1)
#정규성 x
# 정상성 o
# 자기상관성 아직 x

# 잔차진단(아직도 발생한 오차)을 했는데,
# 아직 만족스럽지 않아 => 그렇다는 것은 성능개선을 해야된다는 말임
```

```python
residual2 = m2.resid
residual_diag(residual2)
```

#### ② AIC
* 선형 모델에서의 적합도와, feature가 과도하게 늘어나는 것을 방지하도록 설계된 통계량이 AIC 입니다.
* 값이 작을 수록 좋은 모델
* 공식 : 𝐴𝐼𝐶=−2 ln⁡(𝐿)+2𝑘 ➡ - 모델의 적합도 + 변수의 갯수
* SARIMAX 모델.aic로 쉽게 통계량을 구할 수 있습니다.

```python
m1.aic
```

```python
m2.aic
```

#### ③ Validation

시계열 데이터로 실제값과 예측값에 대해 비교하여 그래프를 그려봅시다.
- forecast로 예측(.fit이 아님)

```python
pred1 = m1.forecast(30)
print("mean absolute error:", mean_absolute_error(y_val, pred1))
print("mean absolute percentage error:", mean_absolute_percentage_error(y_val, pred1))
```

```python
pred2 = m2.forecast(30)
print("mean absolute error:", mean_absolute_error(y_val, pred2))
print("mean absolute percentage error:", mean_absolute_percentage_error(y_val, pred2))
```

* 결과 시각화

```python
plot_model_result(y_train, y_val, pred1)
```

```python
plot_model_result(y_train, y_val, pred2)
```

## (3) 하이퍼파라미터 튜닝

실제로 p, d, q를 찾는 과정은 마치 Grid Search 처럼 값을 조금씩 조정해가며  최적의 모델을 찾아가는 과정과 유사합니다.


### 1) 학습

```python
from itertools import product
```

* 값의 범위 지정
    * 시간이 많이 소요될 수 있으니 범위를 적절하게 지정하세요.

```python
# product 함수를 이용하여 값의 조합을 구성
p = [ 1, 2, 3, 4 ]
q = [ 1, 2, 3, 4 ]
d = [ 0, 1 ]
iter = list(product(p,d,q))
iter
```

* 튜닝

```python
# 
mae, aic = [],[]
for i in iter :
    model_fit = sm.tsa.SARIMAX(y_train, order=(i[0],i[1],i[2])).fit()
    pred = model_fit.forecast(30)
    mae.append( mean_absolute_error(y_val, pred))
    aic.append(model_fit.aic)
    print(i)
```

```python
result = pd.DataFrame({'params(p,d,q)' : iter, 'mae' : mae, 'aic':aic})

display(result.loc[result['mae'] == result.mae.min()])
display(result.loc[result['aic'] == result.aic.min()])
```

```python
# 가장 성능이 좋은 p, d, q 값으로 모델을 생성합니다.
m3 = sm.tsa.SARIMAX(y_train, order=[4, 1, 4]).fit()
```

### 2) 평가

#### ① 잔차진단

* residual_diag

```python
resid3 = m3.resid
residual_diag(resid3)

# 정규성 x
# 정상성 o
# 자기상관성 좀 줄긴했으나 아직 x

# pacf에 주기가 있음 => 이런 주기를 seasonality라고 함. (계절성이라고 번역하지만, 사이클/주기가 맞는 것)
# 어떤 간격의 주기가 있다는 것을 seasonality라고 한다.
# => ARIMA로 해결 안됨.
# => SARIMA로, m값을 줘야 함.

# 주기성이 있게 튀어오르는 값은 대문자 p, q로 해결해야 함.
```

#### ② AIC
* 선형 모델에서의 적합도와, feature가 과도하게 늘어나는 것을 방지하도록 설계된 통계량이 AIC 입니다.
* 값이 작을 수록 좋은 모델
* 공식 : 𝐴𝐼𝐶=−2 ln⁡(𝐿)+2𝑘 ➡ - 모델의 적합도 + 변수의 갯수

```python
m3.aic

# 약간 줄었음
```

#### ③ Validation

```python
pred3 = m3.forecast(30)

print("mean absolute error:", mean_absolute_error(y_val, pred3))
print("mean absolute percentage error:", mean_absolute_percentage_error(y_val, pred3))
```

* 결과 시각화

```python
plot_model_result(y_train, y_val, pred3)
```

## (2) 하이퍼파라미터 튜닝
* 너무 큰 범위를 지정하면 시간이 많이 걸립니다.
    * D는 1로 고정
    * P, Q는 1 ~ 4 사이에서 결정

### 1) 학습

```python
P = [1, 2, 3]
Q = [1, 2, 3]
D = [1]
mae, aic = [],[]
iter = list(product(P,D,Q))

for i in iter :
    model_fit = sm.tsa.SARIMAX(y_train, order=( 4 , 1 , 4 ), seasonal_order=(i[0],i[1],i[2],7)).fit()
    pred = model_fit.forecast(30)
    mae.append( mean_absolute_error(y_val, pred))
    aic.append(model_fit.aic)
    print(i)
```

```python
result = pd.DataFrame({'params(P,D,Q)' : iter, 'mae' : mae, 'aic':aic})

display(result.loc[result['mae'] == result.mae.min()])
display(result.loc[result['aic'] == result.aic.min()])
```
