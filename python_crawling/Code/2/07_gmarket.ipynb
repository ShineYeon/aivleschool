### Gmarket 사이트
- 베스트 상품 200개 수집

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# 1. URL
url = 'http://corners.gmarket.co.kr/Bestsellers'
```

```python
# 2. request > response : HTML
```

```python
response = requests.get(url)
response
```

```python
# 3. HTML(str) > BeautifulSoup Object 
```

```python
doc = BeautifulSoup(response.text, "html.parser")
type(doc)
```

```python
# 4. BeautifulSoup Object > css-selector > DataFrame 
```

```python
elements = doc.select("#gBestWrap > div > div:nth-child(5) > div > ul > li")
elements[0]
```

```python
element = elements[0]
data= {
    "title": element.select_one(".itemname").text,
    "o_price": element.select_one(".o-price").text,
    "s_price": element.select_one(".s-price>strong").text,
    "img": "http:"+element.select_one(".lazy").get("data-original"),
}
data
```

```python

items = []
for element in elements:
    data= {
    "title": element.select_one(".itemname").text,
    "o_price": element.select_one(".o-price").text,
    "s_price": element.select_one(".s-price>strong").text,
    "img": "http:"+element.select_one(".lazy").get("data-original"),
    }
    items.append(data)
    
items[:2]
```
