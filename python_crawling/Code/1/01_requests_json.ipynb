### WebCrawling
- 웹 페이지의 종류 (웹페이지 종류에 따라 크롤링 방식이 달라짐)
    - 정적페이지 : 페이지의 데이터가 변경될 때 URL이 변경됨
        => html포맷 데이터 수집
       ex) 페이지 1에서 2누르면 URL1->URL2 변경, 텍스트도 변경됨
    - 동적페이지 : 페이지의 데이터가 변경될 때 URL이 변경되지 않음 
        => json포맷 데이터 수집
       ex) 더보기 버튼클릭했는데 URL이 URL1에서 바뀌지 않음, 근데 텍스트는 추가됨

- requests package
    - 브라우저의 URL을 입력하면 서버에서 데이터를 다운받아 화면에 출력 : URL => DATA
    - requests 패키지 : URL => DATA

```python
#데이터 수집하기 전에 사용할 패키지가 있음
import requests
import pandas as pd #URL로 data수집하면 수집한 데이터를 판다스의 데이터프레임으로 출력할 것임
```

```python
# 1. 웹서비스를 분석 : 크롬 개발자 도구 => URL 얻기. (이게 가장 어려움)
# m.stock.naver.com 네이버 코스피 국내증시의 일별시세 가져오기
```

```python
#10개씩 가져오는데 4페이지에 있는 데이터를 가져온다 라고 추측가능
url = "https://m.stock.naver.com/api/index/KOSPI/price?pageSize=10&page=4"
        
```

```python
data = response.json() #json 포맷의 문자열인데 list로 바꿔줬음 
type(data), data[:2] #list 안에 dictionary 있음 => dictionary 하나하나가 row data가 됨.
```

```python
#데이터 프레임으로 바꿔주기
df = pd.DataFrame(data)[["localTradedAt", "closePrice"]] #모든 칼럼이 필요한게 아니니까 두개만 가져오기
df
```

```python
# 4. 함수 만들기
# params: pagesize, page
```

```python
def stock_price(pagesize, page, code="KOSPI"):
    url = f'https://m.stock.naver.com/api/index/{code}/price?pageSize={pagesize}&page={page}'
    response = requests.get(url) #url로 요청해서 가져오기
    data = response.json() #가져온 데이터를 json()을 통해 리스트 형태로 변환
    return pd.DataFrame(data)[["localTradedAt", "closePrice"]] #리스트로 변환한걸 가지고 데이터 프레임으로 만드는데, 갖고오고싶은 칼럼만 갖고옴
```

```python
stock_price(30, 2)
```

```python
# 1. 웹서비스 분석: URL
```

```python
url = "https://m.stock.naver.com/api/index/KOSDAQ/price?pageSize=10&page=2"
```

```python
# 2. request(url) => response(json) : JSON(str)
```

```python
response = requests.get(url)
data = response.json()
```

```python
# 3. JSON(str) => list, dict => DataFrame
```

```python
# 4. 함수 만들기
```

```python
#default argument
#code="KOSPI" 부분은 디폴트로, 만약에 매개변수 2개만 주면(code를 안줌) KOSPI로 넣어서 작동
def stock_price(pageSize, page, code="KOSPI"):
    """ This function is crawling stock price from naver webpage.
    Params
    ------
    pagesize : int : one page size
    page : int : page number
    code : str : KOSPI or KOSDAQ
    
    Return
    ------
    type : DataFrame : display date, price columns
    """
    url = f"https://m.stock.naver.com/api/index/{code}/price?pageSize={pageSize}&page={page}"
    response = requests.get(url)
    data = response.json()
    return pd.DataFrame(data)[["localTradedAt", "closePrice"]]
```

```python
#최근 60일치 kospi/kosdaq 데이터
kospi = stock_price(60, 1, "KOSPI")
kosdaq = stock_price(60, 1, "KOSDAQ")
display(kospi)
display(kosdaq)
```

```python
#최근 60일치 원달러 환율 데이터 수집
```

```python
# 1. 웹 서비스 분석 : URL
```

```python
page, pageSize = 1, 60
url = f"https://api.stock.naver.com/marketindex/exchange/FX_USDKRW/prices?page={page}&pageSize={pageSize}"
```

```python
# 2. request(url) => response(JSON) : JSON(str)
```

```python
response = requests.get(url)
```

```python
# 3. JSON(str) => list, dict => DataFrame
```

```python

data = response.json() #list 안에 dictionary가 있다
data
```

```python
usd = pd.DataFrame(data)[["localTradedAt", "closePrice"]]
usd.tail(2)
```

```python
# 데이터분석
# 상관관계분석: 두 데이터 집합 사이에 관계가 있는지 확인하는 분석 방법
# 원달러 환율이 높으면 코스피, 코스닥 주가 지수가 낮다. => 음의 상관관계
# 높높이면 양의 상관관계
```

```python
# 피어슨 상관계수 : df.corr()
# 두 데이터 집합에 대한 피어슨 상관계수를 구한 게 1과 가까울수록 강한 양의 상관관계를 갖는다.
# -1과 가까울수록 강한 음의 상관관계를 갖는다.
# 0과 가까울수록 관계가 없다.
```

```python
# 데이터 전처리
df = kospi.copy() #copy()를 쓰지 않고 그냥 df=kospi 해버리면 kospi도 바뀌어 버림.
df["kosdaq"] = kosdaq["closePrice"]
df["usd"] = usd["closePrice"]
df.rename(columns = {"closePrice": "kospi"}, inplace=True)
df.tail(2)
```

```python
df.dtypes #산술연산해야되는데 문자열데이터임
```

```python
#컬럼의 데이터 타입 변경하기 : str => float
# df[column].apply() : 모든 데이터를 함수(람다함수를 말함)에 대입한 결과를 출력
# df[column]에 있는 모든 데이터들을 각각 data에 넣고, float(data.replace(",", "")) 여기로 넘어감
# replace로 콤마(,) 제거하고 float()로 형변환
df["kospi"] = df["kospi"].apply(lambda data: float(data.replace(",", "")))
```

```python
df["kosdaq"] = df["kosdaq"].apply(lambda data: float(data.replace(",", "")))
df["usd"] = df["usd"].apply(lambda data: float(data.replace(",", "")))
```

```python
df.dtypes
```

```python
df[["kospi", "kosdaq", "usd"]].corr()
```

```python
#kospi - kosdaq : 0.984 : 1과 가까우면 강한 양의 상관관계
#kospi - usd : -0.878 : -1과 가까우면 강한 음의 상관관계

# 이렇게 분석해 보니 원달러 환율이 높아서 kospi가 낮게 평가되어있다 => 코스피 주식을 사는 게 맞다.
```

```python
data1 = [1, 2, 3]
data2 = data1
print(data1, data2) # [1, 2, 3], [1, 2, 3]
data1[1] = 4
print(data1, data2) # [1, 4, 3], [1, 4, 3]


"""
 data2도 같이 바뀌었음
 이게 코드 작성 시 실수가 많이 나오는 부분.
 결과가 왜 이렇게 나오는가.??
 변수 선언 = 메모리에 데이터 저장
 data1 = [1, 2, 3] 실행 시 메모리 안에 [1, 2, 3]을 만든다. 
 그리고 메모리의 주솟값이 연결된다, 그리고 d1은 그 주솟값이 들어있음 '참조한다' 라고 함
 data2 = data1 하면 data2가 data1에 있던 주솟값을 가짐 => 참조하는 주소가 같음
 data1[1]=4 라고 하면 data1과 data2가 참조하는 메모리가 같기 때문에
 data1[1]이나 data2[1]이나 같은 값임
 """
```

```python
data1 = [1, 2, 3]
data2=data1 #주솟값 복사(얕은 복사)
data3=data1.copy() #값 복사(깊은 복사)
print(data1, data2, data3) # [1, 2, 3] [1, 2, 3] [1, 2, 3]

data1[1] = 4
print(data1, data2, data3) #[1, 4, 3] [1, 4, 3] [1, 2, 3]

"""
data1과 data2가 참조하는 주솟값은 같음 => 얕은 복사 (주솟값을 복사)
data3가 copy하면 똑같은 걸 만들어 주소를 참조 => 깊은 복사 (실제 값을 복사)
"""
```

```python
# lambda : 일회성 함수 (실행 시 잠깐 생성되었다가 사라지기 때문)
# 사용하는 이유: 간단한 함수(파라미터를 받아서 바로 리턴하는)를 메모리를 절약하여 사용
# 간단한 함수를 argument에서 구현해서 사용하는 용도로 주로 사용
"""
람다는 임시함수로 메모리를 점유하고 있지 않습니다. 
def 함수명 으로 정의된 함수는 프로그램 종료 때 소멸되기 때문에 일명 글로벌 함수라고도 함니다. 
람다는 함수사용후 즉각 메모리에서 소멸됩니다.
"""
```

```python
#함수 3개 사용, 메모리 3칸 사용
def plus(n1, n2):
    return n1+n2

def minus(n1, n2):
    return n1-n2

def calc(func, n1, n2):
    return func(n1, n2)
```

```python
calc(plus, 1, 2), calc(minus, 1, 2) #plus가 calc의 func로 들어감, 1과 2는 각각 n1, n2로 들어감
```

```python
plus_lambda = lambda n1, n2: n1+n2
calc(plus, 2, 3), plus_lambda(2, 3)

#lambda 파라미터 : 리턴코드
```

```python
#람다를 이용해서 바꾸기
def calc(func, n1, n2):
    return func(n1, n2)

calc(lambda n1, n2: n1+n2, 1, 2), calc(lambda n1, n2: n1-n2, 1, 2)
```
